```
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Part 1 Knn model
knn = KNeighborsClassifier()
params_knn = {'n_neighbors': np.arange(1, 25)}
knn_gs = GridSearchCV(knn, params_knn, cv=11)
knn_gs.fit(X_train, y_train)

knn_best = knn_gs.best_estimator_
print(knn_gs.best_params_)

# Part 2 Random Forest 
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=10)
params_rf = {'n_estimators': [50, 100, 200]}
rf_gs = GridSearchCV(rf, params_rf, cv=5)
rf_gs.fit(X_train, y_train)

rf_best = rf_gs.best_estimator_
print(rf_gs.best_params_)

# Part 3 Naive Bayes
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf.fit(X_train,y_train)
print(clf.score(X_test,y_test))


print('kNN: {}'.format(knn_best.score(X_test, y_test)))
print('RF: {}'.format(rf_best.score(X_test, y_test)))
print('Naive Bayes: {}'.format(clf.score(X_test, y_test)))
from sklearn.ensemble import VotingClassifier

estimatorsbm1=[('Naive Bayes', clf),('RF', rf_best),('kNN', knn_best),]

ensemblebm1 = VotingClassifier(estimatorsbm1, voting='hard')
ensemblebm1.fit(X_train,y_train)
print(ensemblebm1.score(X_test,y_test))


estimatorsbm2=[('RF', rf_best),('kNN', knn_best),('Naive Bayes', clf),]

ensemblebm2 = VotingClassifier(estimatorsbm2, voting='hard')
ensemblebm2.fit(X_train,y_train)
print(ensemblebm2.score(X_test,y_test))


estimatorsbm3=[('RF', rf_best),('kNN', knn_best)]

ensemblebm3 = VotingClassifier(estimatorsbm3, voting='hard')
ensemblebm3.fit(X_train,y_train)
print(ensemblebm3.score(X_test,y_test))


estimatorsbm4=[('kNN', knn_best),('RF', rf_best)]

ensemblebm4 = VotingClassifier(estimatorsbm4, voting='hard')
ensemblebm4.fit(X_train,y_train)
print(ensemblebm4.score(X_test,y_test))

```
